```
def obtener_hora_inicio(rango_horario):
    if pd.isna(rango_horario):
        return None
    hora_inicio = rango_horario.split(" ")[1].split(":")[0]  # Toma la primera hora del rango
    return int(hora_inicio)


ACCIDENTES_2010_2018_DB['HORA_INICIO'] = ACCIDENTES_2010_2018_DB['RANGO HORARIO'].apply(obtener_hora_inicio)

ACCIDENTES_2010_2018_DB['FECHA'] = pd.to_datetime(ACCIDENTES_2010_2018_DB['FECHA']) + pd.to_timedelta(ACCIDENTES_2010_2018_DB['HORA_INICIO'], unit='h')

ACCIDENTES_2010_2018_DB[['FECHA', 'RANGO HORARIO']].head()
```

Crear rangos horarios

```
def extraer_hora_inicio(rango_horario):
    hora_inicio_str = rango_horario.split(' ')[1]
    hora_inicio = pd.to_datetime(hora_inicio_str, format='%H:%M').time()
    return hora_inicio

def asignar_franjas_desde_rango(rango_horario):
    hora_inicio = extraer_hora_inicio(rango_horario)
    if hora_inicio >= pd.to_datetime('06:00').time() and hora_inicio < pd.to_datetime('12:00').time():
        return 'Mañana'
    elif hora_inicio >= pd.to_datetime('12:00').time() and hora_inicio < pd.to_datetime('19:00').time():
        return 'Tarde'
    elif hora_inicio >= pd.to_datetime('19:00').time() and hora_inicio <= pd.to_datetime('23:59').time():
        return 'Noche'
    else:
        return 'Madrugada'

ACCIDENTES_2010_2018_DB['FRANJA HORARIA'] = ACCIDENTES_2010_2018_DB['RANGO HORARIO'].apply(asignar_franjas_desde_rango)
ACCIDENTES_2010_2018_DB.drop('RANGO HORARIO', axis=1, inplace=True)

ACCIDENTES_2010_2018_DB.head()
```

```
def limpiar_texto(texto):

    if pd.isnull(texto):
        return texto
    texto = str(texto)
    texto = re.sub(r', [^,]*$', '', texto)
    texto = re.sub(r'/[^/]*$', '', texto)
    texto = re.sub(r'-[^-]*$', '', texto)
    texto = re.sub(r'\s+[a-zA-Z]\s+', ' ', texto)
    texto = re.sub(r'^[a-zA-Z]\s+', '', texto)
    texto = re.sub('[\.\-]', ' ', texto)
    texto = texto.replace('KM', '')
    texto = texto.strip()
    palabras = texto.split(" ")
    palabras_limpias = [palabra for palabra in palabras if len(palabra) > 1]
    texto_limpio = " ".join(palabras_limpias)
    return texto_limpio


ACCIDENTES_2010_2018_DB['DISTRITO'] = ACCIDENTES_2010_2018_DB['DISTRITO'].apply(limpiar_texto)
ACCIDENTES_2019_2023_DB['DISTRITO'] = ACCIDENTES_2019_2023_DB['DISTRITO'].apply(limpiar_texto)

ACCIDENTES_2010_2018_DB['LUGAR ACCIDENTE'] = ACCIDENTES_2010_2018_DB['LUGAR ACCIDENTE'].apply(limpiar_texto)
ACCIDENTES_2019_2023_DB['LUGAR ACCIDENTE'] = ACCIDENTES_2019_2023_DB['LUGAR ACCIDENTE'].apply(limpiar_texto)
```
